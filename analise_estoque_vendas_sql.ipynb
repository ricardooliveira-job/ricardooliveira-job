```python
import pandas as pd
import kagglehub
from pandasql import sqldf

# 1. Baixar os dados (Garantindo que o caminho exista)
path = kagglehub.dataset_download("olistbr/brazilian-ecommerce")
items = pd.read_csv(f"{path}/olist_order_items_dataset.csv")
products = pd.read_csv(f"{path}/olist_products_dataset.csv")

# 2. Criar a tabela df_prod unindo Itens e Produtos
df_prod = pd.merge(items, products, on='product_id')

# 3. Definir a Query SQL
query = """
SELECT
    product_category_name AS Categoria,
    COUNT(order_item_id) AS Total_Vendido
FROM df_prod
GROUP BY product_category_name
ORDER BY Total_Vendido DESC
LIMIT 10;
"""

# 4. Executar o SQL forçando o Python a olhar para a tabela correta
# Passamos locals() diretamente para o sqldf não se perder
ranking_sql = sqldf(query, locals())

print("--- RANKING LOGÍSTICO GERADO VIA SQL ---")
print(ranking_sql)

```
