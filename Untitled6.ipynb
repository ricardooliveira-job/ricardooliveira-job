```python
import pandas as pd
import kagglehub

# 1. Baixa a versão mais recente do dataset da Olist
path = kagglehub.dataset_download("olistbr/brazilian-ecommerce")

print("Caminho dos arquivos:", path)

# 2. Carrega o arquivo de PEDIDOS (Orders) usando o caminho que o Kagglehub criou
# O path contém vários arquivos, precisamos indicar qual queremos (olist_orders_dataset.csv)
df_orders = pd.read_csv(f"{path}/olist_orders_dataset.csv")

# 3. Converte as colunas de tempo para o formato de DATA (essencial para logística)
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'])
df_orders['order_delivered_customer_date'] = pd.to_datetime(df_orders['order_delivered_customer_date'])

print("\nSucesso! Arquivo carregado via Kagglehub.")
print(f"O dataset tem {df_orders.shape[0]} linhas.")

# Exibe as primeiras linhas para conferir
df_orders.head()
```

```python
import pandas as pd

# 1. Criar a coluna de Lead Time (em dias)
# Subtraímos a data de entrega pela data da compra
df_orders['lead_time_dias'] = (df_orders['order_delivered_customer_date'] - df_orders['order_purchase_timestamp']).dt.days

# 2. Criar a coluna de 'Atraso'
# Comparamos a data de entrega real com a data que foi prometida (estimada)
df_orders['order_estimated_delivery_date'] = pd.to_datetime(df_orders['order_estimated_delivery_date'])
df_orders['atraso_dias'] = (df_orders['order_delivered_customer_date'] - df_orders['order_estimated_delivery_date']).dt.days

# 3. Filtrar apenas os pedidos que já foram entregues para não sujar a média
df_entregues = df_orders.dropna(subset=['lead_time_dias'])

# --- RESULTADOS PARA SEU RELATÓRIO ---
print(f"Média Geral de Entrega: {df_entregues['lead_time_dias'].mean():.1f} dias")
print(f"Maior demora registrada: {df_entregues['lead_time_dias'].max()} dias")

# Quantos pedidos chegaram DEPOIS do prazo prometido?
total_atrasos = df_entregues[df_entregues['atraso_dias'] > 0].shape[0]
percentual_atraso = (total_atrasos / df_entregues.shape[0]) * 100

print(f"\nTaxa de Ineficiência (Atrasos): {percentual_atraso:.2f}%")
```

```python
import pandas as pd
import kagglehub

# 1. Garantir que os dados estão baixados
path = kagglehub.dataset_download("olistbr/brazilian-ecommerce")

# 2. Carregar Orders e Customers
df_orders = pd.read_csv(f"{path}/olist_orders_dataset.csv")
df_customers = pd.read_csv(f"{path}/olist_customers_dataset.csv")

# 3. Unir as tabelas (Merge) para saber o estado de cada pedido
df_logistica = pd.merge(df_orders, df_customers, on='customer_id')

# 4. Converter colunas para Data
df_logistica['order_delivered_customer_date'] = pd.to_datetime(df_logistica['order_delivered_customer_date'])
df_logistica['order_estimated_delivery_date'] = pd.to_datetime(df_logistica['order_estimated_delivery_date'])

# 5. Criar coluna que identifica se houve atraso (1 para sim, 0 para não)
df_logistica['houve_atraso'] = (df_logistica['order_delivered_customer_date'] > df_logistica['order_estimated_delivery_date']).astype(int)

# 6. Agrupar por Estado e calcular a porcentagem de atraso
ranking_atraso = df_logistica.groupby('customer_state')['houve_atraso'].mean() * 100
ranking_atraso = ranking_atraso.sort_values(ascending=False).reset_index()

# 7. Renomear colunas para ficar profissional
ranking_atraso.columns = ['Estado', '% de Atraso']

print("--- RANKING DA PIOR LOGÍSTICA POR ESTADO ---")
print(ranking_atraso.head(10)) # Mostra os 10 estados com mais atrasos
```

```python
import plotly.express as px

# 1. Criar o gráfico de barras interativo
fig = px.bar(
    ranking_atraso,
    x='Estado',
    y='% de Atraso',
    title='Taxa de Atraso na Entrega por Estado (Olist)',
    labels={'% de Atraso': 'Percentual de Atraso (%)', 'Estado': 'UF'},
    color='% de Atraso', # As barras mudam de cor conforme o atraso aumenta
    color_continuous_scale='Reds' # Tons de vermelho para indicar criticidade
)

# 2. Ajustar o layout para ficar mais limpo e profissional
fig.update_layout(
    xaxis_tickangle=-45,
    yaxis_range=[0, ranking_atraso['% de Atraso'].max() + 5],
    template='plotly_white'
)

# 3. Mostrar o gráfico
fig.show()
```

```python
import plotly.express as px

# 1. Carregar a tabela de itens (que tem o freight_value)
df_items = pd.read_csv(f"{path}/olist_order_items_dataset.csv")

# 2. Unir com nossa tabela de logística que já tem o Lead Time e o Estado
df_completo = pd.merge(df_logistica, df_items, on='order_id')

# 3. Agrupar por Estado para ver a relação Frete Médio vs. Prazo Médio
df_relacao = df_completo.groupby('customer_state').agg({
    'freight_value': 'mean',
    'lead_time_dias': 'mean'
}).reset_index()

# 4. Criar um Gráfico de Dispersão (Scatter Plot)
# Ele mostra se existe uma linha de tendência entre preço e velocidade
fig_relacao = px.scatter(
    df_relacao,
    x="freight_value",
    y="lead_time_dias",
    text="customer_state",
    title="Relação: Custo do Frete vs. Tempo de Entrega (por Estado)",
    labels={'freight_value': 'Frete Médio (R$)', 'lead_time_dias': 'Prazo Médio (Dias)'},
    trendline="ols" # Adiciona uma linha de tendência automática
)

fig_relacao.update_traces(textposition='top center')
fig_relacao.show()
```

```python
import pandas as pd
import kagglehub
import plotly.express as px

# 1. Baixar/Localizar os dados
path = kagglehub.dataset_download("olistbr/brazilian-ecommerce")

# 2. Carregar as 3 tabelas necessárias
orders = pd.read_csv(f"{path}/olist_orders_dataset.csv")
items = pd.read_csv(f"{path}/olist_order_items_dataset.csv")
customers = pd.read_csv(f"{path}/olist_customers_dataset.csv")

# 3. Unir tudo em um único DataFrame (O "Geralzão")
df = pd.merge(orders, items, on='order_id')
df = pd.merge(df, customers, on='customer_id')

# 4. Converter datas e calcular o Lead Time (Garantindo que a coluna exista aqui)
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])

# Cálculo do tempo de entrega em dias
df['lead_time_dias'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days

# Remover linhas sem data de entrega para não dar erro no gráfico
df_clean = df.dropna(subset=['lead_time_dias'])

# 5. Agrupar por Estado para o Insight de Negócio
df_relacao = df_clean.groupby('customer_state').agg({
    'freight_value': 'mean',
    'lead_time_dias': 'mean'
}).reset_index()

# 6. Gerar o Gráfico de Dispersão Profissional
fig = px.scatter(
    df_relacao,
    x="freight_value",
    y="lead_time_dias",
    text="customer_state",
    trendline="ols",
    title="Consultoria Logística: Custo de Frete vs. Rapidez de Entrega",
    labels={'freight_value': 'Frete Médio (R$)', 'lead_time_dias': 'Prazo Médio (Dias)'},
    template="plotly_white"
)

fig.update_traces(textposition='top center')
fig.show()

print("Análise concluída com sucesso!")
```

```python
import plotly.express as px

# 1. Criar colunas de Ano e Mês (usando a data da compra)
df_clean['ano_mes'] = df_clean['order_purchase_timestamp'].dt.to_period('M').astype(str)

# 2. Calcular a taxa de atraso média por mês
# Criamos uma coluna booleana (True/False) e tiramos a média (vira percentual)
df_clean['atrasou'] = df_clean['order_delivered_customer_date'] > df_clean['order_estimated_delivery_date']
sazonalidade = df_clean.groupby('ano_mes')['atrasou'].mean().reset_index()
sazonalidade['atrasou'] = sazonalidade['atrasou'] * 100 # Transforma em %

# 3. Gerar Gráfico de Linha (O "Eletrocardiograma" da Logística)
fig_sazonal = px.line(
    sazonalidade,
    x='ano_mes',
    y='atrasou',
    title='Evolução da Taxa de Atraso (Sazonalidade Olist)',
    labels={'ano_mes': 'Mês da Compra', 'atrasou': '% de Pedidos Atrasados'},
    markers=True
)

fig_sazonal.show()
```
